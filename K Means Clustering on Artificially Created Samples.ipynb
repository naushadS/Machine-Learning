# Importing Visualization Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# Creating the Artificial Dataset

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101)
```

```python
X.shape #200 observation has been created with 2 features
```

```python
X[:2] #Sample: Generated Samples of first 2 Observations
```

```python
y.shape
```

```python
y[:2] # Sample: Integer Labels for cluster membership of first 2 Observations 
```

# Visualize data

```python
plt.scatter(X[:,0],X[:,1],c=y)
#[Refer: (input cell marked 3)] 
#X[:,0] means select all the rows but only 0th column
#c parameter is to color code the data points as per their corresopnding integer labels for cluster membership 
```

# Creating the clusters

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
fig , (ax1,ax2) = plt.subplots(1,2,sharey=True,figsize=(10,6))

ax1.set_title('Kmeans')
ax1.scatter(X[:,0],X[:,1],c=kmeans.labels_)

ax2.set_title('Original')
ax2.scatter(X[:,0],X[:,1],c=y)
```

Ignoring the colors of each of the data points, the kmeans model seems to have assigned the same into reasonable clusters.
